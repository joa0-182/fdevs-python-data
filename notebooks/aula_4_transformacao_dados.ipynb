# 4. Criação de colunas derivadas

Colunas derivadas são criadas a partir de colunas que já existem na base.

### Exemplo: valor por unidade


```python
base_vendas["VALOR UNITARIO"] = base_vendas["VALORLIQUIDO"] / base_vendas["VOLUME"]

base_vendas[["NOMEPRODUTO", "VOLUME", "VALORLIQUIDO", "VALOR UNITARIO"]].head()
```

### Exemplo: cópia de uma coluna com outro significado de análise


```python
base_vendas["CLIENTE"] = base_vendas["DESCRICAOCLIENTE"]

base_vendas[["DESCRICAOCLIENTE", "CLIENTE"]].head()
```

# 5. Usando assign()

O **assign()** permite criar colunas novas de forma organizada.


```python
base_vendas = base_vendas.assign(
    VALOR_DOBRADO = base_vendas["VALORLIQUIDO"] * 2,
    VOLUME_DOBRADO = base_vendas["VOLUME"] * 2
)

base_vendas[["VALORLIQUIDO", "VALOR_DOBRADO", "VOLUME", "VOLUME_DOBRADO"]].head()
```

# 6. Usando rename()

O **rename()** serve para renomear colunas da base.


```python
base_renomeada = base_vendas.rename(columns={
    "DATAEMISSAO": "DATA VENDA",
    "VALORLIQUIDO": "VALOR TOTAL LIQUIDO"
})

base_renomeada.head()
```

# 7. Usando map()

O **map()** é útil para traduzir ou classificar valores de uma coluna.


```python
base_vendas[["TIPOFATURAMENTO", "DESCRICAOTIPOFATURAMENTO"]].drop_duplicates()
```

```python
mapa_faturamento = {
    1: "Operação Comercial",
    2: "Ajuste de inventário",
    3: "Cortesia/Amostra",
    13: "Operação Comercial",
    14: "Ajuste de inventário",
   
}

base_vendas["TIPO OPERACAO"] = base_vendas["TIPOFATURAMENTO"].map(mapa_faturamento)

base_vendas[["TIPOFATURAMENTO", "TIPO OPERACAO"]].drop_duplicates()
```

Se algum valor não estiver no dicionário, o resultado será **NaN**.


# 8. Usando apply() com consciência

O **apply()** é útil em casos mais personalizados, mas deve ser usado com cuidado.

Em muitas situações, operações diretas e vetorizadas são melhores.


```python
def classificar_valor(valor):
    if valor >= 1000:
        return "Alto"
    else:
        return "Baixo"

base_vendas["FAIXA_VALOR"] = base_vendas["VALORLIQUIDO"].apply(classificar_valor)

base_vendas[["VALORLIQUIDO", "FAIXA_VALOR"]].head(10)
```

# 9. Criando colunas com condições, utilizando o NumPy

A biblioteca **NumPy** é perfeita para calculos, e bastante performática quando se trata de muitos dados, geralmente é melhor pratica do que usar o **.apply()**

Em muitas situações, operações diretas e vetorizadas são melhores.


#### **np.where** é basicamente um if e else da biblioteca

```python
base_vendas["CLASSIFICACAO VALOR"] = np.where(
    base_vendas["VALORLIQUIDO"] >= 1000, 
    "Alto", 
    "Baixo"
)

base_vendas[["VALORLIQUIDO", "FAIXA_VALOR", "CLASSIFICACAO VALOR"]].head(10)
```

#### **np.select** conseguimos trabalhar com diversas condições

```python
base_vendas["CLASSIFICACAO VALOR COM CONDICOES"] = np.select(
    [
        base_vendas["VALORLIQUIDO"] >= 1000, 
        (base_vendas["VALORLIQUIDO"] < 1000) & (base_vendas["VALORLIQUIDO"] >= 500),
        base_vendas["VALORLIQUIDO"] < 500
    ], 
    [
        "Alto", 
        "Médio", 
        "Baixo"
    ],
    default="Desconecido"
)

base_vendas[["VALORLIQUIDO", "FAIXA_VALOR", "CLASSIFICACAO VALOR", "CLASSIFICACAO VALOR COM CONDICOES"]].head(10)
```

# 10. Manipulação de strings

As funções de string ajudam a padronizar e extrair partes de textos.


### Transformando nome do produto em maiúsculo


```python
base_vendas["NOMEPRODUTO MAIUSCULO"] = base_vendas["NOMEPRODUTO"].str.upper()

base_vendas[["NOMEPRODUTO", "NOMEPRODUTO MAIUSCULO"]].head()
```

### Extraindo a primeira palavra do nome do produto


```python
base_vendas["PRIMEIRA_PALAVRA_PRODUTO"] = base_vendas["NOMEPRODUTO"].str.split().str[0]

base_vendas[["NOMEPRODUTO", "PRIMEIRA_PALAVRA_PRODUTO"]].head()
```

# 11. Conversão de datas

Antes de trabalhar com datas, precisamos garantir que a coluna está no tipo correto.


```python
base_vendas["DATAEMISSAO"] = pd.to_datetime(base_vendas["DATAEMISSAO"])

base_vendas[["DATAEMISSAO"]]
```

```python
base_vendas["DATAEMISSAO"].dtype
```

# 12. Manipulação de datas

Depois da conversão, podemos extrair partes da data.


### Criando colunas de ano, mês e dia


```python
base_vendas["ANO"] = base_vendas["DATAEMISSAO"].dt.year
base_vendas["MES"] = base_vendas["DATAEMISSAO"].dt.month
base_vendas["DIA"] = base_vendas["DATAEMISSAO"].dt.day

base_vendas[["DATAEMISSAO", "ANO", "MES", "DIA"]].head()
```

### Criando o nome do dia da semana


```python
base_vendas["DIA_SEMANA"] = base_vendas["DATAEMISSAO"].dt.day_name()

base_vendas[["DATAEMISSAO", "DIA_SEMANA"]].head()
```

### Criando o nome do dia da semana em português


```python
base_vendas["DIA_SEMANA"] = base_vendas["DATAEMISSAO"].dt.day_name(locale="pt_BR")

base_vendas[["DATAEMISSAO", "DIA_SEMANA"]].sample(5)
```

### Criando uma coluna no formato ano-mês


```python
base_vendas["ANO_MES"] = base_vendas["DATAEMISSAO"].dt.strftime("%Y-%m")

base_vendas[["DATAEMISSAO", "ANO_MES"]].head()
```
