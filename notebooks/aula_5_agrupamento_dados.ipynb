## Métricas de Agregação (Groupby)

### Básicas:
* **count()** -> Calcula o número de valores não nulos em cada grupo.
* **size()** -> Retorna o tamanho total do grupo (incluindo valores nulos).
* **sum()** -> Soma todos os valores do grupo.
* **mean()** -> Calcula a média aritmética simples.
* **median()** -> Retorna a mediana (valor central) do grupo.

### Dispersão e Estátistica:

* **min()** -> Identifica o valor mínimo dentro do grupo.
* **max()** -> Identifica o valor máximo dentro do grupo.
* **std()** -> Calcula o desvio padrão (o quanto os dados se afastam da média).
* **var()** -> Calcula a variância dos dados.
* **quantile(q)** -> Retorna o valor em um determinado quantil (ex: 0.25 para o 1º quartil).

### Seleção e Unidade:

* **nunique()** -> Conta quantos valores distintos (únicos) existem no grupo.
* **first()** -> Retorna o primeiro valor de cada grupo conforme a ordem do índice.
* **last()** -> Retorna o último valor de cada grupo.
* **unique()** -> Retorna uma lista com todos os elementos únicos do grupo.

# 1. Qual cliente mais fatura?


```python
clientes = base.groupby("DESCRICAOCLIENTE")["VALORLIQUIDO"].sum()
clientes.sort_values(ascending=False).head()

clientes.head()
```

# 2. Agrupamento com múltiplas métricas


```python
base.columns
```

```python
clientes = (
    base.groupby("DESCRICAOCLIENTE")
    .agg({
        "VALORLIQUIDO": "sum",
        "VOLUME": "sum"
    })
)

clientes.head()
```

# 3. Múltiplas métricas na mesma coluna


```python
base.groupby("DESCRICAOCLIENTE")["VALORLIQUIDO"].agg(["sum","mean","max"]).head()
```

# 4. Resetando índice


```python
clientes = base.groupby("DESCRICAOCLIENTE")["VALORLIQUIDO"].sum().reset_index()
clientes.head()
```

# 5. Ordenação


```python
clientes.sort_values(by="VALORLIQUIDO", ascending=False).head()
```

# 6. Agrupamento por múltiplas colunas


```python
base.groupby(["DESCRICAOCLIENTE","TIPOFATURAMENTO"])["VALORLIQUIDO"].sum().reset_index().head(10)
```

# 7. Produto que mais gera receita


```python
base.groupby(["NOMEPRODUTO"])["VALORLIQUIDO"].sum().sort_values(ascending=False).reset_index().head()
```
