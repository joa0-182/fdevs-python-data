# 1. Importando o Pandas

Vamos começar importando a biblioteca **Pandas**.


```python
import pandas as pd
```

# 2. Lendo a base de dados

Vamos carregar a base que será usada nos exemplos.


```python
arquivo = r"C:\Users\Joao.Maria\Documents\curso-python_joao_pedro\Base\Base_analitica_vendas.xlsx"

base_vendas = pd.read_excel(arquivo, sheet_name="Base")
```

# 3. Seleção de colunas

Podemos selecionar uma ou mais colunas da base.

### Selecionando uma coluna


```python
base_vendas.columns
```

```python
base_vendas["DESCRICAOCLIENTE"]
```

### Selecionando várias colunas


```python
base_vendas[["DESCRICAOCLIENTE", "NOMEPRODUTO", "VALORLIQUIDO"]]
```

# 4. Seleção com loc

O **loc** seleciona dados usando **rótulos**: nomes de linhas e colunas.

### Selecionando linhas e colunas específicas


```python
base_vendas.loc[234:242, ["DESCRICAOCLIENTE", "DESCRICAOGRUPOPRODUTO", "VOLUME", "VALORLIQUIDO"]]
```

### Selecionando todas as linhas de uma coluna


```python
base_vendas.loc[:, ["DESCRICAOCLIENTE", "VALORLIQUIDO"]]
```

# 5. Seleção com iloc

O **iloc** seleciona dados pela **posição numérica**.

### Selecionando linhas e colunas por posição


```python
base_vendas.iloc[12:15, 2:7]
```

### Selecionando apenas algumas colunas por posição


```python
base_vendas.iloc[:, [2, 6, 10]]
```

# 6. Seleção Bidimensional

Seleção em formato de DataFrame **[ [] ]**


```python
base_vendas[["CODIGOGRUPOPRODUTO", "NOMEPRODUTO", "VALORLIQUIDO"]]
```

# 7. Filtros booleanos

Filtros booleanos servem para retornar apenas as linhas que atendem a uma condição.

### Filtrando valores maiores que 1000


```python
base_vendas[base_vendas["VALORLIQUIDO"] > 1000]
```

### Filtrando um tipo de faturamento específico


```python
base_vendas[base_vendas["DESCRICAOTIPOFATURAMENTO"] == "VENDA"]
```

### Combinando duas condições


```python
base_vendas[(base_vendas["VALORLIQUIDO"] > 1000) & (base_vendas["VOLUME"] > 10)]
```

# 8. Usando query()

O **query()** permite escrever filtros de forma mais próxima de uma expressão textual.


```python
base_vendas.query("VALORLIQUIDO > 1000")
```

Também podemos combinar condições.


```python
base_vendas.query("VALORLIQUIDO > 1000 and VOLUME > 10")
```

Também podemos passar váriaveis


```python
valor_liquido_minimo = 20000
volume_minimo = 45

base_vendas.query(f"VALORLIQUIDO > @valor_liquido_minimo and VOLUME > {volume_minimo}")
```

Também podemos passar váriaveis e selecionar as colunas que queremos usar


```python
valor_liquido_minimo = 20000
volume_minimo = 45

base_vendas.query(f"VALORLIQUIDO > @valor_liquido_minimo and VOLUME > {volume_minimo}")[["DESCRICAOCLIENTE", "DESCRICAOGRUPOPRODUTO"]]
```

# 9. Usando unique()

O **unique()** mostra os valores únicos de uma coluna.


```python
base_vendas["TIPOFATURAMENTO"].unique()
```

# 10. Usando value_counts()

O **value_counts()** conta quantas vezes cada valor aparece.


```python
base_vendas["TIPOFATURAMENTO"].value_counts()
```

Também podemos usar em outras colunas categóricas.


```python
base_vendas["DESCRICAOGRUPOPRODUTO"].value_counts()
```

# 11. Usando sort_values()

O **sort_values()** ordena os dados por uma ou mais colunas.

### Ordenando do menor para o maior valor líquido


```python
base_vendas.sort_values("VALORLIQUIDO")
```

### Ordenando do maior para o menor valor líquido


```python
base_vendas.sort_values("VALORLIQUIDO", ascending=False)
```

### Ordenando por mais de uma coluna


```python
base_vendas.sort_values(["TIPOFATURAMENTO", "VALORLIQUIDO"], ascending=[True, False])
```
