## 2. Carregando os dados

```python
caminho_arquivo_base_venda = r"C:\Users\Joao.Maria\Documents\curso-python_joao_pedro\Base\Captura_venda_marco.json"
caminho_configuracao_produto = r"C:\Users\Joao.Maria\Documents\curso-python_joao_pedro\Base\Configuração Produto.xlsx"
caminho_configuracao_cliente = r"C:\Users\Joao.Maria\Documents\curso-python_joao_pedro\Base\Configuração Cliente.xlsx"
```

## 4. Merge com Cliente (LEFT JOIN)

### Mantém todas as vendas, mesmo sem cliente correspondente.

```python
base = pd.merge(
    base_venda, 
    configuracao_cliente, 
    on='CODIGOCLIENTE', 
    how='left'
)

base.head()
```

#### Tem como também trazer apenas os campos ao qual a gente precisa

```python
base_cliente = pd.merge(
    base_venda, 
    configuracao_cliente[["CODIGOCLIENTE", "DESCRICAOCLIENTE"]], 
    on='CODIGOCLIENTE', 
    how='left'
)

base_cliente.head()
```

## 5. Merge com Produto

```python
base = pd.merge(
    base, 
    configuracao_produto[["CODIGOPRODUTO", "NOMEPRODUTO", "DESCRICAOGRUPOPRODUTO", "DESCRICAOFAMILIA"]], 
    on='CODIGOPRODUTO', 
    how='left'
)

base.head()
```

## 6. Problema real – Chaves faltantes


```python
base.isnull().sum()
```

### Se houver valores nulos em cliente ou produto, o relacionamento falhou.

## 8. Tipos de Ligação

    how=
* **inner** → Mantém apenas os registros que existem nas duas tabelas
* **left** → Mantém todos os registros da tabela da esquerda e traz correspondências da direita
* **right** → Mantém todos os registros da tabela da direita e traz correspondências da esquerda
* **outer** → Mantém todos os registros de ambas as tabelas, combinando onde houver correspondência

```python
clientes_faltantes = pd.merge(
    base_venda,
    configuracao_cliente[["CODIGOCLIENTE", "DESCRICAOCLIENTE"]],
    on="CODIGOCLIENTE",
    how="right"
)
```

### Filtrando apenas os clientes faltantes

```python
clientes_faltantes = clientes_faltantes[clientes_faltantes["DATAEMISSAO"].isnull()]

clientes_faltantes.head()
```

### Identificando os valores nulos de ambas as bases

#### indicator=True cria uma coluna:

* left_only -> só existe em vendas
* right_only -> só existe em clientes
* both -> existe nos dois

```python
comparacao = pd.merge(
    base_venda,
    configuracao_cliente[["CODIGOCLIENTE", "DESCRICAOCLIENTE"]],
    on="CODIGOCLIENTE",
    how="outer",
    indicator=True
)

comparacao.head(25)
```
