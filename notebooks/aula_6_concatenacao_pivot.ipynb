## 2. Carregando os dados
Vamos carregar três bases:
- Dezembro
- Janeiro
- Fevereiro

```python
caminho_arquivo = r"C:\Users\Joao.Maria\Documents\Analise\Base\Base_analitica_vendas_periodo.xlsx"
```

```python
base_venda_dezembro = pd.read_excel(caminho_arquivo, sheet_name="Dez")
base_venda_janeiro = pd.read_excel(caminho_arquivo, sheet_name="Jan")
base_venda_fevereiro = pd.read_excel(caminho_arquivo, sheet_name="Fev")
```

## 3. Analisando estrutura das bases

## 4. Concatenando os datasets

```python
base_venda_dezembro["MES"] = "Dezembro"
base_venda_janeiro["MES"] = "Janeiro"
base_venda_fevereiro["MES"] = "Fevereiro"
```

```python
base_vendas_periodo = pd.concat([
    base_venda_dezembro, 
    base_venda_janeiro, 
    base_venda_fevereiro
], ignore_index=True)
```

```python
base_vendas_periodo
```

```python
base_vendas_periodo.shape
```

## 5. Padrozinando os campos antes de concatenar

```python
base_venda_dezembro = base_venda_dezembro.rename(columns={
    "DATA": "DATAEMISSAO"
})
```

```python
base_vendas_periodo = pd.concat([
    base_venda_dezembro, 
    base_venda_janeiro, 
    base_venda_fevereiro
], ignore_index=True)
```

## 6. Verificando o resultado

```python
base_vendas_periodo.shape
```

## 8. Pivot

Agora vamos transformar a base consolidada em uma tabela analítica, facilitando a leitura dos dados.

### Pivot por Mês

Quanto foi faturado em cada mês?

```python
pivot_mes = base_vendas_periodo.pivot_table(
    values="VALORLIQUIDO",
    index="MES",
    aggfunc="sum"
)

display(pivot_mes)
```

### Por Mês e Tipo Faturamento

Como o faturamento se distribui por tipo?

```python
pivot_faturamento = base_vendas_periodo.pivot_table(
    values="VALORLIQUIDO",
    index="MES",
    columns="TIPOFATURAMENTO",
    aggfunc="sum",
    fill_value=0
)

pivot_faturamento
```

### Faturamento por familia de produto e mês

Como o faturamento se distribui por tipo?

```python
pivot_grupo = base_vendas_periodo.pivot_table(
    values="VALORLIQUIDO",
    index="DESCRICAOGRUPOPRODUTO",
    columns="MES",
    aggfunc="sum",
    fill_value=0
)

pivot_grupo
```

### Sem ordenação, a tabela fica inútil.

```python
pivot_grupo = pivot_grupo.sort_values(
    by="Fevereiro",
    ascending=False
)

pivot_grupo
```

### Múltiplas métricas

```python
pivot_multi = base_vendas_periodo.pivot_table(
    values=["VALORLIQUIDO", "VOLUME"],
    index="MES",
    aggfunc="sum"
)

pivot_multi
```
