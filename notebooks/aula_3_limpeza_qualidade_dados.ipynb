# Aula 03 – Limpeza e Qualidade de Dados

## Objetivo da aula

Nesta aula vamos aprender como tratar dados para garantir qualidade na análise.

**Você irá aprender:**
- copy()
- isna()
- fillna()
- dropna()
- duplicated()
- drop_duplicates()
- Padronização de strings
- Conversão de tipos
- Remoção de colunas desnecessárias

## 1. Importando o Pandas

```python
import pandas as pd
```

## 2. Lendo a base de dados

```python
arquivo = r"C:\Users\Joao.Maria\Documents\curso-python_joao_pedro\Base\Base venda bruta marco.csv"
base_vendas = pd.read_csv(arquivo, sep=";")
```

## 3. Visualizando a base

```python
base_vendas.head()
```

```python
base_vendas.info()
```

## 4. Criando uma cópia da base

```python
base_vendas_tratado = base_vendas.copy()
```

## 5. Identificando valores nulos

```python
base_vendas_tratado.isna().sum()
```

## 6. Tratando valores nulos

```python
base_vendas_tratado["VOLUMESELLOUT"] = base_vendas_tratado["VOLUMESELLOUT"].fillna(0)

base_vendas_tratado[["VOLUMESELLOUT"]].sample(6)
```

```python
base_vendas_tratado["UF"] = base_vendas_tratado["UF"].fillna("DESCONHECIDO")

base_vendas_tratado[["UF"]].sample(6)
```

## 7. Removendo registros inválidos

```python
base_vendas_tratado = base_vendas_tratado.dropna(subset=["DATA"])

base_vendas_tratado[["DATA"]].isnull().sum()
```

## 8. Verificando duplicidade

```python
base_vendas_tratado.duplicated()
```

## 9. Removendo duplicados

```python
base_vendas_tratado = base_vendas_tratado.drop_duplicates()

base_vendas_tratado.duplicated()
```

## 10. Padronização de strings

```python
base_vendas_tratado["UF"] = base_vendas_tratado["UF"].str.strip().str.upper()

base_vendas_tratado["UF"].sample(6)
```

```python
base_vendas_tratado["NOMEPRODUTO"] = base_vendas_tratado["NOMEPRODUTO"].str.strip()

base_vendas_tratado.head(3)
```

## 11. Conversão de tipos

```python
base_vendas_tratado["DATA"] = pd.to_datetime(base_vendas_tratado["DATA"])

base_vendas_tratado.info()
```

```python
base_vendas_tratado["PRECO"] = base_vendas_tratado["PRECO"].astype(float)

base_vendas_tratado.info()
```

## 12. Removendo coluna desnecessária

```python
base_vendas_tratado = base_vendas_tratado.drop(columns=["DATACARGA"])
```

## 13. Conferindo resultado final

```python
base_vendas_tratado.head()
```

```python
base_vendas_tratado.info()
```
